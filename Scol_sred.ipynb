```python
alpha = 0.3
alpha1 = 0.05
y_mod_a = [0];
y_mod_a1 = [0];
for sa in range(len(y)):
    if y_mod_a[-1] == 0:
        y_mod_a.append(alpha*y[sa]+(1-alpha)*y[sa])
        y_mod_a1.append(alpha1*y[sa]+(1-alpha1)*y[sa])
    else:
        y_mod_a.append(alpha*y[sa]+(1-alpha)*y_mod_a[-1])
        y_mod_a1.append(alpha*y[sa]+(1-alpha1)*y_mod_a1[-1])
print(y_mod_a[-1])
print(y_mod_a1[-1]) # при alpha = 0.05

# print(y_e,y_e1)
```
